```python
import os 
import pandas as pd
import datetime
import numpy as np
import matplotlib.pyplot as plt
# import subprocess #ベクター画像出力用
import matplotlib.dates as mdates # DataTimeの横軸調整
```

```python
csv_file_name = os.listdir('input')
csv_file_name

trigger_times = []
for filename in csv_file_name:
    filepath = 'input/' + filename
    print(filepath,type(filepath))
    df = pd.read_csv(filepath, names=range(32)) # 行によってカラム数が違うとエラーを吐くのでnamesで列名を指定する
    trigger_time = (df.loc[df.iloc[:,0] == "Trigger Time"]).iloc[0,1] # Trigger Timeの値を検索して代入
    trigger_time = trigger_time.replace("'", "20")
    trigger_time = trigger_time[0:19]
    dt_trigger_time = datetime.datetime.strptime(trigger_time, "%Y-%m-%d %H:%M:%S")
    trigger_times.append(dt_trigger_time)

def time_change(x):
    dt2 = dt1 + datetime.timedelta(seconds=x)
    return dt2

count = 0
for (filename,trigger_time) in zip(csv_file_name, trigger_times):
    count += 1
    filepath = 'input/' + filename
    df1 = pd.read_csv(filepath, header=11)
    dt1 = trigger_time
    df1['Time'] = df1['Time'].apply(time_change)
    df1 = df1.set_index('Time', drop=True) # 引数drop=Falseとすると、指定列がindexに設定された上でデータ列にもそのまま残る
    # print(df1)
    if count == 1:
        df = df1
    if count >= 2:
        df = pd.concat([df, df1])

df = df.resample('s').interpolate() # 線形補完
df = df.resample('10min').interpolate() # 線形補完

df = df.where(df <= 100, -10)

df = df.drop(df.columns[list(range(0,15))], axis=1) # 有線データの削除
df = df.drop(df.columns[-1], axis=1) # 有線データの削除
df.to_csv('output/temp.csv')
df
```
